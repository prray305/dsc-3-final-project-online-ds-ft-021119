The comments should be cleaned for better consistency in Bag of Words, especially through removing punctiations.

```python
def clean_comment(comment):
    cleaned_comment = []
    tokenized = tokenize(comment)
    for sentence in tokenized:
        for word in sentence:
#             if not '<' in word and not '>' in word:
                if not 'br/' in word:
                    for symbol in ",.?''<>":
                        word = word.replace(symbol, '').lower()
                    cleaned_comment.append(word.strip())
    return cleaned_comment
```

```python
def count_vectorize(comment, vocab=None):
    if vocab:
        unique_words = vocab
    else:
        unique_words = list(set(comment))
    
    comment_dict = {i:0 for i in unique_words}
    
    for word in comment:
        comment_dict[word] +=1
        
    return comment_dict
```

```python
def term_frequency(BoW_dict):
    total_word_count = sum(BoW_dict.values())
    
    for ind, val in BoW_dict.items():
        BoW_dict[ind] = val / total_word_count
        
    return BoW_dict
```

```python
#calculate inverse document frequency
def inverse_document_frequency(list_of_dicts):
    vocab_set = set()
    for d in list_of_dicts:
        for word in d.keys():
            vocab_set.add(word)
    
    full_vocab_dict = {i:0 for i in vocab_set}
    
    for word, val in full_vocab_dict.items():
        docs = 0
        
        for d in list_of_dicts:
            if word in d:
                docs += 1
        
        full_vocab_dict[word] = np.log((len(list_of_dicts)/float(docs)))
    
    return full_vocab_dict
```

```python
print(inverse_document_frequency(term_freq_test_comment))
```

```python
def tf_idf(list_of_dicts):
    doc_tf_idf = {}
    idf = inverse_document_frequency(list_of_dicts)
    full_vocab_list = {i:0 for i in list(idf.keys())}
    
    tf_idf_list_of_dicts = []
    
    for doc in list_of_dicts:
        doc_tf = term_frequency(doc)
        for word in doc_tf:
            doc_tf_idf[word] = doc_tf[word] * idf[word]
        tf_idf_list_of_dicts.append(doc_tf_idf)
        
    return tf_idf_list_of_dicts
```

```python
def main(data):
    cleaned_comments = []
    count_vectorized_all_documents =[]
    for comment in data:
        cleaned = clean_comment(comment)
        cleaned_comments.append(cleaned)
#         count_vectorized_comments = count_vectorize(cleaned)
    count_vectorized_comments = count_vectorize(cleaned_comments)
    count_vectorized_all_documents.append(count_vectorized_comments)
    
    tf_idf_all_docs = tf_idf(count_vectorized_all_documents)
    return count_vectorized_all_documents
#     return tf_idf_all_docs   #change back to previous format
```
